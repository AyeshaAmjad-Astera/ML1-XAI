```python
import dice_ml
import numpy as np
import json
from catboost import CatBoostClassifier
from sklearn.metrics import f1_score, roc_auc_score, accuracy_score
```

```python
import sys
sys.path.append("../")
import pandas as pd
from utils.data_utils import DataLoader
```

```python
with open('C:/Users/ayesha.amjad/Documents/GitHub/MSDS-ML-Project/ML1-XAI/Project/models/CatBoost/best_param_enc2.json', 'r') as file:
    best_params = json.load(file)
```

```python
data_loader = DataLoader(file_path='../data/raw/Churn_Modelling.xls' , target_column='Exited', test_size=0.2, random_state=42, clean_data=True)
```

```python
df = data_loader.encoded_data
df.info()
```

```python
X_train, y_train = data_loader.get_train_data_enc()
X_test, y_test = data_loader.get_test_data_enc()
#y_train
```

```python
model = CatBoostClassifier(**best_params)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(f"F1 Score {f1_score(y_test, y_pred, average='macro')}")
print(f"Accuracy {accuracy_score(y_test, y_pred)}")
print(f"ROC_AUC {roc_auc_score(y_test, y_pred)}")
```

```python
Concat_df = pd.concat([X_train, y_train], axis=1)
Concat_df
```

```python
Concat_df['NumOfProducts'] = Concat_df['NumOfProducts'].astype(float)
Concat_df['IsActiveMember'] = Concat_df['IsActiveMember'].astype(float)
```

```python
data_dice = dice_ml.Data(dataframe=Concat_df, 
                         # For perturbation strategy
                         continuous_features=['Age', 
                                              'NumOfProducts',
                                              'EstimatedSalary',
                                              'IsActiveMember'], 
                         outcome_name='Exited')
# Model
model_dice = dice_ml.Model(model=model, 
                        # There exist backends for tf, torch, ...
                        backend="sklearn")
explainer = dice_ml.Dice(data_dice, 
                         model_dice, 
                         # Random sampling, genetic algorithm, kd-tree,...
                         method="random")
```

```python
# Generate CF based on the blackbox model
input_datapoint = X_test[0:1]
#input_datapoint['NumOfProducts'] = input_datapoint['NumOfProducts'].astype(float)
#input_datapoint['IsActiveMember'] = input_datapoint['IsActiveMember'].astype(float)

input_datapoint
```

```python
cf = explainer.generate_counterfactuals(input_datapoint, 
                                  total_CFs=4, 
                                  permitted_range=None,
                                  desired_class="opposite")
# Visualize it
cf.visualize_as_dataframe(show_only_changes=True)
```

```python
features_to_vary=['Age', 
                  'NumOfProducts',
                  'EstimatedSalary',
                  'IsActiveMember']
permitted_range={'Age':[20,30],
                 'EstimatedSalary':[116000, 119000]}
# Now generating explanations using the new feature weights
cf = explainer.generate_counterfactuals(input_datapoint, 
                                  total_CFs=34, 
                                  desired_class="opposite",
                                  permitted_range=permitted_range,
                                  features_to_vary=features_to_vary)
# Visualize it
cf.visualize_as_dataframe(show_only_changes=True)
```
